```python
element = 'oxygen'
print('first three characters:', element[0:3])
print('last three characters:', element[3:6])
```

```python
element = 'oxygen'
print('first four characters:', element[:4])
print('last two characters:', element[4:])
print('last character:', element[-1])
print('before last character:', element[-2])
print('element name:', element[:])
```

```python
import matplotlib.pyplot as plt
x = [3,2,5,4,5]
y = [2,5,9,8,12]
plt.plot(x, y)
plt.xlabel("Numbers")
plt.ylabel("Even numbers")
plt.title("Numbers")
plt.show()
```

```python
import matplotlib.pyplot as plt

height = [150, 160,165,170,175,180,185]
weight = [50, 55, 60, 65, 70, 75, 80]

plt.scatter(height, weight)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.show()
```
