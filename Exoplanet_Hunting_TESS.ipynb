```python
!pip install lightkurve --quiet
!pip install lightkurve

import lightkurve as lk
import matplotlib.pyplot as plot

TIC = 'TIC 172900988'
sector_data = lk.search_lightcurve(TIC, mission='TESS')
sector_data

if len(sector_data) == 0:
    print(f"No data found for TIC {TIC}. Try another TIC ID.")
else:
    lc = sector_data[0].download().remove_nans()

    lc.plot(
       linewidth=0,
       marker='.',
       color='midnightblue',
       alpha=1
    )

    c_norm = lc.normalize()
    c_flat = c_norm.flatten(window_length=401)

    c_flat.plot(
       linewidth=0,
       marker='.',
       alpha=1
    )
```
